```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.indeed.com/jobs?q=Data%20Science&l&vjk=f47d225eb0a02f2c")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
dejob = soup.find(id= 'resultsBodyContent')

dejob.prettify()

detitle = [deti.get_text() for deti in dejob.select(".tapItem .jobTitle")]

desdesc = [desd.get_text() for desd in dejob.select(".tapItem .job-snippet")]

desal = [desa.get_text() for desa in dejob.select(".tapItem .salaryOnly")]

deloc = [delo.get_text() for delo in dejob.select(".tapItem .companyLocation")]

deconame = [deco.get_text() for deco in dejob.select(".tapItem .companyName")]
```

```python
print(len('Title'), len('Short Description'), len('Salary')),len('Location'),len('Company Name')
```

```python
a = {'Title' : detitle ,'Short Description' : desdesc , 'Salary': desal , 'Location':deloc , 'Company' : deconame}
datasciencejobs = pd.DataFrame.from_dict(a, orient='index')
datasciencejobs= datasciencejobs.transpose()
```

```python
datasciencejobs
```

```python
datasciencejobs['Short Description'] =datasciencejobs['Short Description'].str.replace('\n', '')
```

```python
"""Save as .csv file"""
datasciencejobs.to_csv("datasciencejobs.csv", index=False)
```
